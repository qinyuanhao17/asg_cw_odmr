```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Num = 1001#Number of frequency points
StartFreq = 1 # In GHz
EndFreq = 2 # In GHz
RFLevel = 20.0 # In dBm 
RFTime = 100.0 # In ms
RFFreqOffset = 0.0  # In Hz
RFLevelOffset = 0.0  # In dB

```

```python
SN = list(range(2,Num+2,1))
length = len(SN)

freq_float = np.linspace(StartFreq,EndFreq,length)
freq_str = []
for i in range(length):
    item = str(round(freq_float[i],3)) + ' GHz'
    freq_str.append(item)

level_float = SN.copy()
level_str = []
for i in range(length):
    item = str(round(RFLevel,3))+ ' dBm'
    level_str.append(item)

time_float = SN.copy()
time_str = []
for i in range(length):
    item = str(round(RFTime,3)) + ' ms'
    time_str.append(item)

FreqOffset_float = SN.copy()
FreqOffset_str = []
for i in range(length):
    item = str(round(RFFreqOffset,3)) + 'Hz'
    FreqOffset_str.append(item)

LevelOffset_float = SN.copy()
LevelOffset_str = []
for i in range(length):
    item = str(round(RFLevelOffset,3)) + ' dB'
    LevelOffset_str.append(item)

init_csv = pd.DataFrame([np.array([1, '9 kHz', '-110.0 dBm', '500.0 ms', '0.0 Hz', '0.0 dB'])], columns= ['SN', 'Freq', 'Level', 'Time', 'Freq Offset', 'Level Offset'])

RFList = np.array([SN, freq_str,level_str, time_str, LevelOffset_str, FreqOffset_str]).transpose()

RF_CSV = pd.DataFrame(RFList, columns= ['SN', 'Freq', 'Level', 'Time', 'Freq Offset', 'Level Offset'])
RF_CSV[['SN']] = RF_CSV[['SN']].astype(int)
RF_CSV.index += 1
RF_CSV
```

```python
shuffled_rf_csv = RF_CSV.sample(frac = 1)
```

```python

final_csv = pd.concat([init_csv, shuffled_rf_csv],axis=0)
final_csv.to_csv('sweep_list_analysis.csv')
final_csv.to_csv('sweep_list.csv', index=None)
```
